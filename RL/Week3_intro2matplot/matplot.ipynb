```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 12, 1000)
y = np.cos(x)

plt.plot(x, y)
plt.title("line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.linspace(0, 12, 20)
y = np.cos(x)

plt.scatter(x, y, marker="x", c=["red"])
plt.title("Scatter")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, edgecolor='black', linewidth=0.5, color="blue")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 9))

x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(1000)

axes[0][0].plot(x, y1, color="red")
axes[1][0].set_title("Line")
axes[1][0].set_xlabel("X-axis")
axes[1][0].set_ylabel("Y-axis")
axes[0][1].scatter(x, y2)
axes[1][0].set_title("Scatter")
axes[1][0].set_xlabel("X-axis")
axes[1][0].set_ylabel("Y-axis")
axes[1][0].hist(data, color="green", edgecolor='black', linewidth=0.5)
axes[1][0].set_title("Histogram")
axes[1][0].set_xlabel("Value")
axes[1][0].set_ylabel("Frequency")
axes[1][1].bar(["apple", "pen", "pineapple"], [10, 20, 30], color="purple", label=["apple", "pen", "pineapple"])
axes[1][1].set_title("Bar")
axes[1][1].set_ylabel("Values")

plt.show()
```
